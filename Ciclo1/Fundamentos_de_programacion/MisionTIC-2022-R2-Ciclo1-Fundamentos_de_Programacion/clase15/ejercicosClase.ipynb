```python
import matplotlib.pyplot as plt
x1=[1,3,4,5,8]
y1=[0.5,2.5,3,5,6]
x2=[0.5,1,2,4,7]
y2=[0.5,1,2,4,7]

plt.plot(x1,y1, color='green', linewidth=5, label='Experimental')

plt.plot(x2,y2, color='black', linewidth=3,label='Teorico')

plt.grid()#para ver la grilla
plt.title('Datos teoricos vs Experimental')#para ponerle un titulo
plt.xlabel('Eje X')#poner un nombre a el eje x
plt.ylabel('Eje Y')#poner un nombre a el eje y


```

```python
valores=[20,40,60,80]
#la suma de todos es el 100% 
plt.pie(valores, autopct='%1.1f%%', colors=['red','blue','yellow','orange','violet'],labels=['Prekinder', 'Kinder','primaria','secundaria'], shadow=True, startangle=90, explode=(0.1,0,0.2,0))
plt.title('Grafica torta')
plt.show()
```

```python
import pandas as pd
df = pd.read_csv('titanic3.csv')
df.head()
```

```python
cant=df.survived.value_counts()
#cant=df.survived.value_counts(normalize=True) no me muestra las 800 perssonas que murieron
print(cant)
cant.plot(kind='bar',alpha=1)
```

```python
plt.figure(figsize=(30,20))
#barra kde hace una linea sobre las barras 
for class_i in [1,2,3]:
    df.age[df.pclass== class_i].plot(kind='kde')
```

```python
df.plot(kind='bar',x='age',y='pclass',color='red')
plt.show()
```

```python
plt.figure(figsize=(30,20))
#barra kde hace una linea sobre las barras 
for age_i in [1,2,3]:
    df.age[df.pclass== age_i].plot(kind='bar')
```

```python
#edades en eun histograma
r=range(round(min(df.age)),round(max(df.age)+1))
plt.hist(df.age,r,histtype='bar')
plt.show()
```
